# Crear data set


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.metrics import adjusted_rand_score
```

```python
data = datasets.load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
data
```

```python
X = data.data
y = data.target
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=1 )
kmeans.fit(X)
```

```python
labels = kmeans.fit_predict(X)
print(labels)
```

# Visualization

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6, edgecolors='k')
plt.title('Clusters K-Means en el dataset IRIS (Características originales)')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.show()
```

# To get 2 componets

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca
```

## To make graphics data

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = labels, cmap='viridis', edgecolors='k')
plt.title('Clusters K-Means en el dataset IRIS (PCA - 2D)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid()
plt.show()
```

```python
inercia = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=1 )
    kmeans.fit(X)
    inercia.append(kmeans.inertia_)
    print(inercia)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(range(1,11),inercia, c='black', edgecolors='k')
plt.title('IRIS (PCA - 2D)')
plt.xlabel('k')
plt.ylabel('Inercia')
plt.plot(range(1,11),inercia)
```
